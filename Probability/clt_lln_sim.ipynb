**Central Limit Theorem, Law of Large Numbers visualization**

```python
from scipy.stats import randint, uniform, expon
import matplotlib.pyplot as plt
```

Our plan is to illustrate the CLT by taking an increasing number of samples 
independently from the same distribution, and examining the (frequency) 
distribution of the standardized sum.

This means taking a number of samples for EACH sum.


```python
# parameters for whatever distribution

# this is a discrete uniform ranging over the values 0, 1, 2, ..., 99, 100
a, b = 0, 100
dist = randint
# take care in defining your distribution, getting its parameters correct!
# check scipy.stats docs for details.

def X(a, b):
    return dist(a, b)

def S(n):
    return sum(X(a,b).rvs(n))

mu = dist.mean(a,b) # mean of X
std = dist.std(a,b) # std  of X
```

```python
# standardized sum:
# use this one to visualize Central Limit Theorem
def Z(n):
    return (S(n) - n * mu) / (std * n**0.5)

# sample mean:
# use this one to visualize Law of Large Numbers
def X_bar(n):
    return S(n) / n
```

To see the Central Limit Theorem in action, take repeated sample sums of a distibution. 

Increase the number of samples (n) in each sum, and *standardize* the sum by 
 - subtracting off n * the population mean (to make the mean of the sum zero), and 
 - divide by sqrt(n) * population standard deviation (to make the standard deviation of the sum one).

(Examine the function Z(n) above to see what I mean. Notice that this is the same process we use to get a z-score from an x-score, but for a sum of x's instead of just one x.)

Notice that the distribution in the histograms becomes standard normal as the number of sample sums increases.

```python
# Central Limit Theorem:
# The distribution of Z(n) converges to a standard normal as n increases.
# We'll display this fact by sampling Z(n) and plotting histograms as n increases.
    
num_sample_sums = 1000

fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(10,10))

print("Central Limit Theorem")
samples_per_sum_list = [1, 5, 10, 50, 100, 500, 1000, 5000, 10000]
for n, samples_per_sum in enumerate(samples_per_sum_list):
    row = n // 3
    col = n  % 3
    ax = axes[row][col]

    print(f"Finished with {samples_per_sum} samples per sum for {num_sample_sums} sample sums.")
    sample_sums = [Z(samples_per_sum) for i in range(num_sample_sums)]
    ax.hist(sample_sums)
    ax.set_title(f"Z({samples_per_sum}): {num_sample_sums} samples")
```

To see the Law of Large Numbers in action, take repeated sample means of a distibution.
As you increase the number of samples (n) in the sample means, the sample variance shrinks, 
and the sample mean distribution becomes normal. Since the sample variance is shrinking, 
the sample mean distribution converges to the population mean.

Notice that the distribution in the histograms is normal as the number of sample means increases, 
and that the variance is decreasing (the x-axis numbers are getting closer and closer together.)

```python
# Law of Large Numbers:
# The distribution of X_bar(n) converges to the population mean as n increases.
# We'll display this fact by sampling X_bar(n) and plotting histograms as n increases.
    
num_sample_sums = 1000

fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(10,10))

print("Law of Large Numbers")
samples_per_sum_list = [1, 5, 10, 50, 100, 500, 1000, 5000, 10000]
for n, samples_per_sum in enumerate(samples_per_sum_list):
    row = n // 3
    col = n  % 3
    ax = axes[row][col]

    print(f"Finished with {samples_per_sum} samples per sum for {num_sample_sums} sample sums.")
    sample_sums = [X_bar(samples_per_sum) for i in range(num_sample_sums)]
    ax.hist(sample_sums)
    ax.set_title(f"X_bar({samples_per_sum}): {num_sample_sums} samples")
    
```
