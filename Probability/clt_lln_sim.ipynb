**Central Limit Theorem visualization**

```python
from scipy.stats import randint, uniform, expon
import matplotlib.pyplot as plt
```

Our plan is to illustrate the CLT by taking an increasing number of samples 
independently from the same distribution, and examining the (frequency) 
distribution of the standardized sum.

This means taking a number of samples for EACH sum.


```python
# parameters for whatever distribution
a, b = 1, 26

# take care in defining your distribution, getting its parameters correct!
# check scipy.stats docs for details.

dist = randint

def X(a, b):
    return dist(a, b)

def S(n):
    return sum(X(a,b).rvs(n))

mu = dist.mean(a,b) # mean of X
std = dist.std(a,b) # std  of X
```

```python
# standardized sum:
# use this one to visualize Central Limit Theorem
def Z(n):
    return (S(n) - n * mu) / (std * n**0.5)

# sample average:
# use this one to visualize Law of Large Numbers
def X_bar(n):
    return S(n) / n
```

```python
# Central Limit Theorem:
# The distribution of Z(n) converges to a standard normal as n increases.
# We'll display this fact by sampling Z(n) and plotting histograms as n increases.
    
num_sample_sums = 1000

fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(10,10))

print("Central Limit Theorem")
samples_per_sum_list = [1, 5, 10, 50, 100, 500, 1000, 5000, 10000]
for n, samples_per_sum in enumerate(samples_per_sum_list):
    row = n // 3
    col = n  % 3
    ax = axes[row][col]

    print(f"Finished with {samples_per_sum} samples per sum for {num_sample_sums} sample sums.")
    sample_sums = [Z(samples_per_sum) for i in range(num_sample_sums)]
    ax.hist(sample_sums)
    ax.set_title(f"Z({samples_per_sum}): {num_sample_sums} samples")
```

```python
# Law of Large Numbers:
# The distribution of X_bar(n) converges to the population mean as n increases.
# We'll display this fact by sampling X_bar(n) and plotting histograms as n increases.
    
num_sample_sums = 1000

fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(10,10))

print("Law of Large Numbers")
samples_per_sum_list = [1, 5, 10, 50, 100, 500, 1000, 5000, 10000]
for n, samples_per_sum in enumerate(samples_per_sum_list):
    row = n // 3
    col = n  % 3
    ax = axes[row][col]

    print(f"Finished with {samples_per_sum} samples per sum for {num_sample_sums} sample sums.")
    sample_sums = [X_bar(samples_per_sum) for i in range(num_sample_sums)]
    ax.hist(sample_sums)
    ax.set_title(f"X_bar({samples_per_sum}): {num_sample_sums} samples")
```
